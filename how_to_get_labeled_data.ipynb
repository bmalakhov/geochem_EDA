### How to get labeled data from Scott Halley's geochemistry dataset

```python
import sqlite3
import pandas as pd

#Scott Halley's geochemistry data in geo3.db file
ls_db_file_name = 'geo3.db' # if geo3.db is in default notebook folder

#an sql file with query that retrieves only clear data labeled as exclusion_ref = 0 
ls_sql_file_name = 'get_clear_data.sql' # if how_to_get_clear_data.sql is in default notebook folder, you can modify it as simple txt file

#create empty dataframe
df = pd.DataFrame(None)

#create database connection
connect_db = sqlite3.connect(ls_db_file_name)

#extract sql string from file ls_sql_file_name
ls_sql1 = open(ls_sql_file_name, 'r').read() 

# ! Sometime sqlite engine does not contain math functions like pow(x,y) etc. and you cannot use it in sql query. 
# It depends on the version of sqlite3.dll
# To calculate concentration, use python function pow(mant, expo)
```

```python
#this retrives data from db to dataframe
query = connect_db.execute(ls_sql1)
cols = [column[0] for column in query.description]
df = pd.DataFrame.from_records(data = query.fetchall(), columns = cols)
```
